```python
df_sub.head()
```

### What is our linear regression equation?

$$ Y = 675 + 39.03x_4 + 70.08x_5 + 70.20x_6 + 73.71x_7$$

### Let's summarize our findings!

```python
"""
We have an r2 score of .81, which is nice and we have 
high confidence because all of our pvalues are 0.000
feature_4 is contributing the least to our predictions, because it's 
coefficient has the smallest magnitude while features 5 through 7 are
contributing similarly with feature_7 having the greatest weight of 73.71.

Skewness and Kurtosis are 0 and 3.0 respectively which implies that our 
predictions are normally distributed.

JB score is low -> normal
"""
```

### Let's write out our equation as a function and calculate the error by hand!

```python
def y_hat(x, params):
    intercept = params[0]
    coeffs = params[1:]
    y = intercept + np.dot(coeffs, x)
    return y
```

```python
x_vals = df_sub.drop(["target"], axis=1).values
x_vals
```

```python
y_vals = np.array([y_hat(x_i, linreg.params) for x_i in x_vals])
y_vals
```

```python
def RMSE(y_true, y_pred):
    num = np.sum((y_true - y_pred)**2)
    den = np.sum((y_true - y_true.mean())**2)
    return 1 - num/den
```

```python
RMSE(df["target"], y_vals)
```

```python
residuals = df_sub.target - y_vals
```

```python
plt.figure(figsize=(6, 6))
plt.scatter(np.linspace(0, 1, 10000), residuals)
plt.hlines(y=0, xmin=0, xmax=1, colors='r') 
plt.ylim(-2, 2)
plt.show()
```

```python
plt.hist(residuals, bins=20)
plt.show()
```
